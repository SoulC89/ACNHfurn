```python
from bs4 import *
import requests as rq
import os
```

```python
r2 = rq.get("https://gamewith.net/animal-crossing-new-horizons/article/show/17525")
soup2 = BeautifulSoup(r2.text, "html.parser")

links = []
finks = []

imager = soup2.select('img[src^="https://gamewith-en.akamaized.net/article_tools/animal-crossing-new-horizons/gacha/"]')

for img in imager:
    links.append(img['src'])
    finks.append(img['alt'])
#for l in links:
#    print(l)
```

```python
import pandas as pd
import numpy as np
```

```python
os.mkdir('acnh_photos_new')
i = 1

for index, img_link in enumerate(links):
    if i <= 5000:
        img_data = rq.get(img_link).content
        with open("acnh_photos_new\\" + str(index + 1) + '.png', 'wb+') as f: 
            f.write(img_data)
        i +=1
    else:
        f.close()
        break
```

```python
from tkinter import Tcl
rlist = Tcl().call('lsort','-dict',mlistnewer)
```

```python
mynewlist = list(rlist)
```

```python
workbook = xlsxwriter.Workbook(r'\Users\Numu\Desktop\Dataisfun\acnh_furnimg.xlsx')
worksheet = workbook.add_worksheet()

worksheet.set_column('B1:B5', 7)
worksheet.set_default_row(45)
image_row = 1 
image_col = 2

for image in mynewlist:
        worksheet.insert_image(image_row,
                          image_col,
                          image,
                          {'x_scale': 0.2,'y_scale' : 0.2,
                          'x_offset':5,'y_offset':5,
                          'positioning' : 1})
        image_row += 1


titles = finks

title_row = 1
title_col = 0 
for title in titles:
    worksheet.write(title_row,
                   title_col,
                   title)
    title_row += 1

workbook.close()
```
